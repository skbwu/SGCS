```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys, copy, os, shutil
import pickle
from tqdm.notebook import tqdm
from IPython.display import clear_output
import seaborn as sns
import torch
    
# friendly colors
colors = ['#377eb8', '#ff7f00', '#4daf4a',
          '#f781bf', '#a65628', '#984ea3',
          '#999999', '#e41a1c', '#dede00']
```

# Class-Specific

```python
# create a dictionary to store our cossim matrices, indexed by {dataset, split, difficulty, class}, idx}
cossim_matrices = {}

# counter for the number of completed figures
counter = 0

# iterate thru each dataset
for dataset in ["CIFAR10", "MNIST", "FashionMNIST"]:
    
    # create a directory for this dataset
    if dataset not in os.listdir("figures/quantitative"):
        os.mkdir(f"figures/quantitative/{dataset}")
        
    # create figures for train/test + each level of difficulty
    for split in ["train", 'test']:
        for difficulty in ["easiest", "hardest", "overall"]:
            
            # create our figure
            fig, ax = plt.subplots(2, 5, figsize=(22, 10))
            
            # create a subdirectory for this {dataset, split, difficulty} combination
            cossim_matrices[(dataset, split, difficulty)] = {}
            
            # iterate thru our classes
            for class_label in tqdm(range(10)):
                
                # another subdirectory for the label
                cossim_matrices[(dataset, split, difficulty)][class_label] = {}
                
                # get the relevant files, each corresponding to an idx with this class label!
                fnames = sorted([f for f in os.listdir(f"maps/{dataset}") if f"difficulty={difficulty}" in f 
                                 and f"split={split}" in f and f"class-label={class_label}" in f])
                
                # iterate thru each of the filenames
                for fname in fnames:
                    
                    # extract the idx of this image
                    idx = int(fname.split("_idx=")[1])
                    
                    # load the .pickle containing the saliency maps for this image
                    with open(f"maps/{dataset}/{fname}", "rb") as file:
                        maps = pickle.load(file)
                        model_descs = list(maps.keys())
                        for model_desc in model_descs:
                            maps[model_desc] = np.array([m.flatten() for m in maps[model_desc]])
                            
                    # construct the cossim matrix for this observation
                    cossim_matrix = np.zeros(shape=(len(model_descs), len(model_descs)))
                    
                    # assemble each entry in our cossim_matrix
                    for i, model_desc1 in enumerate(model_descs):
                        for j, model_desc2 in enumerate(model_descs):

                            # compute Cosine Similarity using dot-product, and then divide out the norms
                            cossim_ij = maps[model_desc1] @ maps[model_desc2].T
                            cossim_ij /= np.sqrt((maps[model_desc1] ** 2).sum(axis=1)).reshape(-1, +1)
                            cossim_ij /= np.sqrt((maps[model_desc2] ** 2).sum(axis=1)).reshape(+1, -1)

                            # special case if i = j, need to ignore the diagonal entries
                            if i == j:
                                np.fill_diagonal(a=cossim_ij, val=np.nan)

                            # flatten, plus keep non-nan values
                            cossim_ij = cossim_ij.flatten()
                            cossim_ij = cossim_ij[~np.isnan(cossim_ij)]

                            # record in our matrix
                            cossim_matrix[i,j] = cossim_ij.mean()

                    # add it to our collection
                    cossim_matrices[(dataset, split, difficulty)][class_label][idx] = cossim_matrix.copy()
                
                # take the mean of all these as our representative + visualize
                cossim_mean = np.mean(list(cossim_matrices[(dataset, split, difficulty)][class_label].values()), axis=0)
                size = 9 if dataset != "CIFAR10" else 13
                sns.heatmap(cossim_mean, ax=ax[class_label // 5, class_label % 5], 
                            fmt=".2f", cbar=False, annot=True, annot_kws = {"size" : size})
                
                # what names to put for our models
                model_strs = []
                for model_desc in model_descs:
                    model_name = model_desc.split("_")[0].upper()
                    model_variant = int(model_desc.split("=")[1].replace("k", ""))
                    model_strs.append(f"{model_name}-{model_variant}")
                
                # labeling + selective labeling
                title_size = 18 if dataset != "CIFAR10" else 25
                ax[class_label // 5, class_label % 5].set_title(f"Class {class_label}", fontsize=title_size)
                if (class_label // 5) == 0:
                    ax[class_label // 5, class_label % 5].set_xticklabels([])
                if (class_label % 5) == 0:
                    fontsize = 12 if dataset != "CIFAR10" else 18
                    ax[class_label // 5, class_label % 5].set_yticklabels(model_strs, rotation=0, 
                                                                          fontsize=fontsize)
                else:
                    ax[class_label // 5, class_label % 5].set_yticklabels([])
                if (class_label // 5) == 1:
                    fontsize = 12 if dataset != "CIFAR10" else 18
                    ax[class_label // 5, class_label % 5].set_xticklabels(model_strs, rotation=90, 
                                                                          fontsize=fontsize)
            
            # beautify at the end
            plt.tight_layout()
            fig_name = f"mean-class-cossims_split={split}_difficulty={difficulty}.png"
            plt.savefig(f"figures/quantitative/{dataset}/{fig_name}", facecolor="white", bbox_inches="tight")
            plt.close()
            
            # status update
            counter += 1
            clear_output(wait=True)
            print(f"Finished generating Figure {counter} of 18.")
```

# Across Classes

```python
# first set of dictionaries: {dataset, split, difficulty} --> {class_label} --> {idx}
for dataset in ["MNIST", "FashionMNIST", "CIFAR10"]:
    
    # make our model strings
    if dataset == "CIFAR10":
        model_strs = ["CNN-25", "CNN-47", "CNN-100", "RESNET-20", "RESNET-32", "RESNET-44"]
    else:
        model_strs = ["MLP-1", "MLP-2", "MLP-3", "CNN-2", "CNN-3", "CNN-4", 
                      "RESNET-20", "RESNET-32", "RESNET-44"]
    
    # generate separate figures for train vs. test splits
    for split in ["train", "test"]:
        
        # create our 1x3 figure (easiest, hardest, overall)
        fig, ax = plt.subplots(1, 3, dpi=200, figsize=(20, 7.5))
        
        # take the MEANS OF THE MATRICES ACROSS ALL CLASSES!
        for i, difficulty in enumerate(["easiest", "hardest", "overall"]):
            
            # compute the mean cossim matrix!
            across_class_means = np.mean([np.mean(list(cossim_matrices[(dataset, split, difficulty)]\
                                                       [clab].values()), axis=0)
                                          for clab in range(10)], axis=0)
            
            # plot our heatmap + label accordingly
            size = 20 if dataset == "CIFAR10" else 15
            sns.heatmap(across_class_means, ax=ax[i], fmt=".2f", cbar=False, annot=True, annot_kws = {"size" : size})
            ax[i].set_title(f"{difficulty.capitalize()} Data Points", fontsize=28)
            ax[i].set_xticklabels(model_strs, rotation=90, fontsize=20)
            if i == 0:
                ax[i].set_yticklabels(model_strs, rotation=0, fontsize=20)
            else:
                ax[i].set_yticklabels([])
                
        # beautify at the end
        plt.tight_layout()
        plt.savefig(f"figures/quantitative/{dataset}/mean-cossims_split={split}_all-difficulties.png",
                    facecolor="white", bbox_inches="tight")
        plt.show()
```
